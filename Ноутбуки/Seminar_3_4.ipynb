### Условный выбор 

При обращении с таблицей нам часто приходится задавать некорые условия.

Например, предположим, что нас интересуют вина, произведенные в Италии, качество которых выше среднего.

Мы можем начать с проверки того, является ли каждое вино итальянским или нет:

```python
reviews.country == 'Italy'
```

Эта операция вывела Series из булевых значений True/False. Этот результат может быть использован внутри __loc__ для выбора соответствующих данных:

```python
reviews.loc[reviews.country == 'Italy']
```

Эта таблица содержит ~20 000 строк. В оригинале было ~130 000. Это означает, что около 15% вин производится в Италии.

Мы также хотели знать, какие из них лучше средних. Вина оцениваются по шкале от 80 до 100 баллов. Пусть нас интересуют вина, набравшие не менее 90 баллов.

Мы можем использовать амперсанд (&), чтобы объединить два условия:

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

Если бы мы хотели выбрать вина, произведенные в Италии __ИЛИ__ имеющие рейтинг выше среднего, мы бы использовали знак __|__:

```python
# объединение множеств

reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

Pandas имеет несколько встроенных условных селекторов.

1) __isin__ позволяет выбирать данные, значение которых находится в списке значений. Например, вот как мы можем использовать __isin__ для выбора вин только из Италии или Франции:

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

2) __null__ (и его спутник __notnull__). Эти методы позволяют выделить значения, которые являются (или не являются) пустыми (NaN). Например, чтобы отфильтровать вина, у которых отсутствует ценник, мы могли бы сделать:

```python
reviews.loc[reviews.price.notnull()]
```

### Map и apply

Используются для преобразования данных.

__map()__ возвращает новый преобразованный Series, который получается в результате применения lambda-функции:

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

__apply()__ действует аналогично, но у него есть также возможность возвращать весь DataFrame:

```python
review_points_mean = reviews.points.mean()
reviews.points.apply(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

Если бы мы вызвали __reviews.apply()__ с __axis='index'__, то вместо передачи функции для преобразования каждой строки нам нужно было бы передать функцию для преобразования каждого столбца.

Обратите внимание, что __map()__ и __apply()__ возвращают новые Series/DataFrame. 
Они __НЕ__ изменяют исходные данные. 

Если мы снова посмотрим на таблицу reviews, то увидим, что она по-прежнему имеет свои первоначальные значения:

```python
reviews.head(2)
```

Для этих целей можно также использовать встроенные операции. Они работают быстрее, чем __map()__ или __apply()__, однако не такие гибкие, поскольку не позволяют выполнять более сложные задачи и функции.

```python
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

Такие операции можно выволнять с Series одинаковой длины. Например, ниже простой способ объединения информации о стране и регионе:

```python
reviews.country + " - " + reviews.region_1
```

### Группировка groupby()

```python
reviews.groupby('points').points.count()
```

```python
# value_counts() - shortcut этой операции 

reviews.points.value_counts()#.sort_index()
```

```python
reviews.groupby('points').price.min()
```

__agg()__ в __groupby()__ позволяет одновременно запускать множество различных функций. Например, мы можем сгенерировать простую статистическую сводку следующим образом:

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

### Сортировка

Можно было заметить, что группировка __groupby()__ возвращает данные согласно порядку индексов (т.е. автоматически происходит сортировка по индексу). Удобнее, чтобы после группировки выводились отсортированные по возрастанию/убыванию данные. Для этого и используют метод __sort_values()__.

```python
countries_reviewed = reviews.groupby(['country']).price.agg([len])
countries_reviewed = countries_reviewed.reset_index()
```

```python
# по умолчанию стоит ascending sort - сортировка по возрастанию

countries_reviewed.sort_values(by='len')
```

```python
# descending sort - сортировка по убыванию

countries_reviewed.sort_values(by='len', ascending=False)
```

Можно выполнять сортировку более чем по одному столбцу одновременно:

```python
countries_reviewed.sort_values(by=['country', 'len'])
```

Метод __sort_index()__ - сортировка по индексу:

```python
countries_reviewed.sort_values(by='len').sort_index()
```

### Добавление/удаление колонки

Можно добавить новую колонку путем присвоения ей константного значения или любой последовательности, имеющей такую же длину, что и таблица. Также можно создать новую колонку с помощью преобразований имеющейся в таблице колонки/колонок, например:

```python
reviews['points_square'] = reviews['points']**2
```

```python
reviews[['points','points_square']].head()
```

Чтобы удалить колонку, используют __drop__:

```python
reviews.drop('points_square', axis=1).head(1)
```

Однако этот метод не переприсваивает таблицу. Видим, что колонка points_square осталась:

```python
reviews.head(1)
```

Для полного удаления можно использовать __inplace = True__ внутри __drop__:

```python
reviews.drop('points_square', axis=1, inplace = True)
```

```python
reviews.head(1)
```

Или же использовать __del__:

```python
# снова создаем колонку

reviews['points_square'] = reviews['points']**2
```

```python
del reviews['points_square']
```

```python
reviews.head(1)
```
